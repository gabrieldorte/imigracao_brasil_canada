```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px  #para criar gráficos interativos
df = pd.read_csv('/content/imigrantes_canada.csv') # Le o arquivo
df.set_index('País', inplace = True) #coloca a coluna pais como index
anos = list(map(str, range(1980, 2014))) #função map converte todos os valores da lista, nesse caso em str
brasil = df.loc['Brasil', anos] #procura o termo brasil e seleciona as imigrações apenas dos anos da variavel anos
brasil_dict = {'Anos': brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()} #cria um dicionario com o index dos anos dos dados do brasil e com os valores de imigração
dados_brasil = pd.DataFrame(brasil_dict) #faz um dataFrame do dicionario
```

```python
fig, ax = plt.subplots(figsize=(8,4)) #cria uma figura para visualizacao
ax.plot(dados_brasil['Anos'], dados_brasil['Imigrantes']) #define os eixos
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013') #o \n muda a linha
ax.set_xlabel('Ano')
ax.set_ylabel('Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #arruma os dados do eixo x
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5)) #1 refere a quantas linhas eu quero nesse subplot, o 2 quantos subplots

axs[0].plot(dados_brasil['Anos'], dados_brasil['Imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Anos')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['Imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3) #ajusta as distancias dos graficos
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')
ymin = 0
ymax = 7000 #determina os valores máximos do Y dos gráficos

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de Imigrantes')

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax) #configura os minimos e máximos do Y

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4)) #cria uma figura para visualizacao
ax.plot(dados_brasil['Anos'], dados_brasil['Imigrantes'], lw=3, color='g') #define os eixos o lw mexe na grossura da linha
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left') #o \n muda a linha o loc diz a posição do titulo
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12) #o labelsize diz o tamanho dos rotulos
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #arruma os dados do eixo x
plt.grid(linestyle='--') #estilo da grade
plt.show()
```

```python
america_do_sul = df.query('Região == "América do Sul"')
america_do_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'grey', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))
ax.bar(america_do_sul.index, america_do_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc = 'left', fontsize = 18)
ax.set_xlabel('')
ax.set_ylabel('Número de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
america_do_sul_ordenada = america_do_sul.sort_values('Total', ascending = True)
cores = []
for pais in america_do_sul_ordenada.index:
  if pais == 'Brasil':
    cores.append('green')
  else:
    cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_do_sul_ordenada.index, america_do_sul_ordenada['Total'], color=cores)
ax.set_title('Brasil foi o quarto país com mais imigrantes no período de\n1980 a 2013', loc = 'left', fontsize = 18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
for i, v in enumerate(america_do_sul_ordenada['Total']):
  ax.text(v, i, int(v), color='black', fontsize=10, ha='left', va='center')  #coloca os valores de cada linha
ax.set_frame_on(False) #tirar a borda do gráfico
ax.get_xaxis().set_visible(False) #tira o eixo x do gráfico
ax.tick_params(axis='both', which='both', length=0) # o length que fala q o tamanho vai ser 0 para ele não aparecer
fig.savefig('america_do_sul_para_o_canada.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

```python
def gerar_grafico_paleta(palette):
  # Creating fig and ax using subplots and setting figsize
  fig, ax = plt.subplots(figsize=(8,4))
  ax = sns.barplot(data=top_10, y = top_10.index, x = 'Total', orient = 'h', palette = palette)
  ax.set_title('Países com maiores imigrações para o Canadá\n no período de 1980 a 2013', loc = 'left', fontsize = 18)
  ax.set_xlabel('Número de imigrantes', fontsize = 18)
  ax.set_ylabel('')
  plt.show()
```
